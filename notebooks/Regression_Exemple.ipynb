```python
""" Paramètres génériques pour tout le portefeuille """
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   
import yfinance as yf             # permet d'utiliser toutes les bases de données de YahooFinance
from scipy import stats
from scipy.optimize import minimize
from IPython.display import display, Math

plt.rcParams["figure.figsize"] = (8,5)   # Dimensions de figure génériques pour tous les graphiques
def resume(nom, valeur, unite=""):
    if isinstance(valeur, float):
        print(f"{nom} : {valeur:.4f} {unite}")
    else:
        print(f"{nom} :\n{valeur}\n")
```

```python
nintendo = yf.download(
    "7974.T",
    start="2019-01-01",
    end="2026-01-31",
    auto_adjust=True
)["Close"]

market = yf.download(
    "^GSPC",
    start="2019-01-01",
    end="2026-01-31",
    auto_adjust=True
)["Close"]

rename_map = {
    "7974.T": "Nintendo",
    "^GSPC": "Market",
}

nintendo = nintendo.rename(columns=rename_map)
market = market.rename(columns=rename_map)
```

```python
nintendo_returns = nintendo.pct_change(fill_method=None).dropna()
market_returns = market.pct_change(fill_method=None).dropna()
nintendo_returns.name = "Nintendo"
market_returns.name = "Market"
```

```python
capm_nintendo = pd.concat([nintendo_returns, market_returns],axis=1).dropna()
print(capm_nintendo.head())
```

```python
y = capm_nintendo["Nintendo"].values
x = capm_nintendo["Market"].values
beta_n, alpha_n, r, p, err = stats.linregress(x, y)

print("===== CAPM NINTENDO =====")
print("Beta :", round(beta_n,3))
print("Alpha journalier :", round(alpha_n,6))
print("R² :", round(r**2,3))
```

```python
alpha_n_annual = alpha_n * 252
print("Alpha annuel (%) :", round(alpha_n_annual * 100,2))
```
