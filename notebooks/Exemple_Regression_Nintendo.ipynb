```python
""" Paramètres génériques pour tout le portefeuille """
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   
import yfinance as yf             # permet d'utiliser toutes les bases de données de YahooFinance
from scipy import stats
from scipy.optimize import minimize
from IPython.display import display, Math

plt.rcParams["figure.figsize"] = (8,5)   # Dimensions de figure génériques pour tous les graphiques
def resume(nom, valeur, unite=""):
    if isinstance(valeur, float):
        print(f"{nom} : {valeur:.4f} {unite}")
    else:
        print(f"{nom} :\n{valeur}\n")
```

```python
nintendo = (
    yf.download(
        "7974.T",
        start="2019-01-01",
        end="2026-01-31",
        auto_adjust=True
    )["Close"]
    .squeeze()
    .rename("Nintendo")
)

market = (
    yf.download(
        "^GSPC",
        start="2019-01-01",
        end="2026-01-31",
        auto_adjust=True
    )["Close"]
    .squeeze()
    .rename("Market")
)
```

```python
# Rendements journaliers
nintendo_returns = nintendo.pct_change(fill_method=None).dropna()
market_returns = market.pct_change(fill_method=None).dropna()
```

```python
# Ajustement temporel
capm_nintendo = pd.concat(
    [nintendo_returns.rename("Nintendo"),
     market_returns.rename("Market")],
    axis=1
).dropna()
```

```python
# Régression
y = capm_nintendo["Nintendo"].values
x = capm_nintendo["Market"].values

beta_n, alpha_n, r, p, err = stats.linregress(x, y)

print("===== CAPM NINTENDO =====")
print("Beta :", round(beta_n,3))
print("Alpha journalier :", round(alpha_n,6))
print("R² :", round(r**2,3))
print("Alpha annuel (%) :", round(alpha_n * 252 * 100,2))
```
